```python
#import libraries
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from textblob import TextBlob
```

```python
# Load the dataset
df = pd.read_csv('Twitter_Reviews.csv')
```

```python
#Print first 5 rows of dataset
df.head()
```

```python
# Tokenization
df['Tokens'] = df['Comment'].apply(lambda x: word_tokenize(x))
```

```python
# POS tagging
df['POS_tags'] = df['Tokens'].apply(lambda x: pos_tag(x))
```

```python
# Sentiment analysis
df['Sentiment'] = df['Comment'].apply(lambda x: TextBlob(x).sentiment.polarity)
```

```python
# Function to classify sentiment based on polarity score
def classify_sentiment(score):
    if score > 0:
        return 'Positive'
    elif score < 0:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
# Classify sentiment using polarity score
df['Sentiment_Classification'] = df['Sentiment'].apply(classify_sentiment)
```
